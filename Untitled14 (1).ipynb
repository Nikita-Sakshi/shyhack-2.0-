```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
df1 = pd.read_csv('C:\\Users\\hp world\\Downloads\\sentiment_statisticscc1e57a.csv') 

df1=df1.dropna()

df1.head()
```

```python
# Load the data
df2 = pd.read_csv('C:\\Users\\hp world\\Downloads\\callsf0d4f5a.csv')

df2=df2.dropna()

# Handle potential datetime parsing errors
for col in ['call_start_datetime', 'call_end_datetime', 'agent_assigned_datetime']:
    try:
      df2[col] = pd.to_datetime(df2[col])
    except pd.errors.ParserError:
      print(f"Warning: Errors encountered while parsing '{col}' column. Check data format.")
      # You can handle parsing errors here (e.g., fill missing values or remove rows)
        
df2['waiting_time'] = (df2['agent_assigned_datetime']-df2['call_start_datetime']).dt.total_seconds()
df2['handle_time'] = (df2['call_end_datetime']-df2['agent_assigned_datetime']).dt.total_seconds()
df2.head()
```

```python
# Finding correlation between columns
print('Correlation between silence_percent_average and average_sentiment : ',df['silence_percent_average'].corr(df['average_sentiment']))
print('Correlation between waiting_time and average_sentiment : ',df['waiting_time'].corr(df['average_sentiment']))
print('Correlation between handle_time and average_sentiment : ',df['handle_time'].corr(df['average_sentiment']))
```

```python
# Plot graph between average sentiment and handle time
xpoints = df['average_sentiment']
ypoints = df['handle_time']

fig, ax = plt.subplots(figsize=(8, 8))
plt.title('Correlation between Sentiment and Handle Time')
plt.xlabel('Average Sentiment')
plt.ylabel('Handle Time (Seconds)')
plt.scatter(xpoints, ypoints)
plt.show()

print('Correlation between average sentiment and average call duration :',df['average_sentiment'].corr(df['handle_time']))
```

```python
#query 6
# giving weightage to each primary call reason using count encoding since no priority is given
# Count encoding is a technique where each category is replaced with the count of its occurrences in the dataset.

count_map1 = df4['primary_call_reason'].value_counts().to_dict()
df4['primary_call_reason_CountEncoded'] = df4['primary_call_reason'].map(count_map1)\

new_df6=(sqldf('''select * from df4 order by primary_call_reason_CountEncoded'''))
new_df6.head(5)
```

```python
# giving weightage to each primary call reason using count encoding since no priority is given

new_df6['primary_call_reason_Encode'] = pd.factorize(new_df6['primary_call_reason'])[0]

agent_tone_order = ['polite', 'angry', 'frustrated', 'calm','neutral']
new_df6['agent_tone_OrdinalEncoded'] = new_df6['agent_tone'].apply(lambda x: agent_tone_order.index(x))

customer_tone_order = ['polite', 'frustrated','angry', 'calm','neutral']
new_df6['customer_tone_OrdinalEncoded'] = new_df6['customer_tone'].apply(lambda x: customer_tone_order.index(x))

new_df6.head(5)
```

```python
# predict primary call reson based on 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import math


x=new_df6[['agent_tone_OrdinalEncoded','waiting_time','primary_call_reason_Encode','customer_tone_OrdinalEncoded']].values
y=new_df6['handle_time'].values

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.27)
#make model
model=LinearRegression()
model.fit(x_train,y_train)

print("Theta Zero or Intercept : ",model.intercept_)
print("Theta One or Slope : ",model.coef_)
r_sq=model.score(x_train,y_train)
print(f"Coefficient of determination : {r_sq}")
y_pred=model.predict(x_test) #direct formula
print("Y Predicted",y_pred)
print('Y actual',y_test)
```
